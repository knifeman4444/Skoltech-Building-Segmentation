```python
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
import random

from Utils.visualization import visualize, get_overlay
```

```python
img = cv2.imread('./data/train/images/train_image_000.png')
mask = cv2.imread('./data/train/masks/train_mask_000.png')
print(img.shape, mask.shape)

EXAMPLE_SIZE = 512
x0 = random.randint(0, img.shape[0] - EXAMPLE_SIZE)
y0 = random.randint(0, img.shape[1] - EXAMPLE_SIZE)
img = img[x0:x0+EXAMPLE_SIZE, y0:y0+EXAMPLE_SIZE]
mask = mask[x0:x0+EXAMPLE_SIZE, y0:y0+EXAMPLE_SIZE]
mask = mask[:, :, 0]
```

```python
visualize(img=img, mask=mask.astype(np.float32))
```

```python
overlay = get_overlay(img, mask)
visualize(overlay=overlay, figsize=(5, 5))
```
